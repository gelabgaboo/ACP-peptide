```python
from google.colab import drive
drive.mount('/content/drive')

# import codecademylib3_seaborn
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CodeCademy Project/tennis_ace_starting/tennis_stats.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 5))
fig.tight_layout()
axes[0,0].scatter(df['FirstServe'], df['FirstServePointsWon'])
axes[0,0].set_title('First Serve vs First Serve Points Won')

axes[0,1].scatter(df['Aces'], df['FirstServePointsWon'])
axes[0,1].set_title('Aces vs First Serve Points Won')

axes[1,0].scatter(df['FirstServePointsWon'], df['Wins'])
axes[1,0].set_title('First Serve Won vs Wins')

axes[1,1].scatter(df['Aces'], df['Wins'])
axes[1,1].set_title('Aces vs Wins')

axes[1,1].scatter(df['Winnings'], df['Ranking'])
axes[1,1].set_title('Winnings vs Ranking')

axes[2,0].scatter(df['BreakPointsOpportunities'], df['Winnings'])
axes[2,0].set_title('Break Points Opportunities vs Winnings')

axes[2,1].scatter(df['BreakPointsOpportunities'], df['Ranking'])
axes[2,1].set_title('Break Points Opportunities vs Ranking')
```

```python
feature = df[['FirstServePointsWon']]
outcome = df[['Wins']]

feature_train, feature_test, outcome_train, outcome_test = train_test_split(feature, outcome, train_size = .8)
model = LinearRegression()
model.fit(feature_train, outcome_train)
model.score(feature_test, outcome_test)
prediction  = model.predict(feature_test)
plt.scatter(outcome_test, prediction, alpha = .4)
```
